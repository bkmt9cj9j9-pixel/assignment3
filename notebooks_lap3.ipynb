```python
plt.figure()
plt.hist(df['Age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.hist(df['Daily_Phone_Hours'])
plt.title("Daily Phone Usage Distribution")
plt.xlabel("Daily Phone Hours")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.scatter(df['Daily_Phone_Hours'], df['Work_Productivity_Score'])
plt.title("Phone Usage vs Productivity")
plt.xlabel("Daily Phone Hours")
plt.ylabel("Productivity Score")
plt.show()
```

```python
plt.figure()
plt.hist(df['Stress_Level'])
plt.title("Stress Level Distribution")
plt.xlabel("Stress Level")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.imshow(df.corr(numeric_only=True))
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()
```
